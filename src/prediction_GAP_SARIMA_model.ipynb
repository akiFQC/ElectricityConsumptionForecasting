## 今回は2007年１月のデータだけを使います。

```python
span_whole = pd.date_range(start=dt.datetime(2007, 1, 1), end=dt.datetime(2007, 3, 1), freq="T")
span_train =pd.date_range(start=dt.datetime(2007, 1, 1), end=dt.datetime(2007, 2, 1), freq="T")
span_test =pd.date_range(start=dt.datetime(2007, 1, 21), end=dt.datetime(2007, 3, 1), freq="T")
```

```python
type(span_whole)
```

```python
ts_gap = df["Global_active_power"][ span_whole]
```

```python
plt.figure(figsize=(16, 4))
ts_gap.plot()
plt.show()
```

```python
sns.distplot(ts_gap)
plt.show()
```

```python
len(ts_gap)
```

## Box-cox変換

```python
ts_gap_bct= stats.boxcox(ts_gap)
```

```python
print(len(ts_gap_bct[0]))
```

```python
ts_gap_bct = pd.Series(ts_gap_bct[0], index=span_whole)
```

```python
ts_gap_bct.head()
```

```python
plt.figure(figsize=(16, 4))
ts_gap_bct.plot()
plt.show()
```

# ARMA modelによる予測

```python
ts_gap_bct[span_train].size
```

```python
model = ARMA(ts_gap_bct[span_train].values, dates= span_train, freq="T", order=(10, 10)).fit(ic="aic")
```

```python
60*24*30
```

```python
pred_ar = model.predict(start=str(span_test[0]), 
                        end=str(span_test[-1]), 
                           dynamic=False)
```

```python
len(pred_ar)
```

```python
print(len(pred_ar))
print(len(span_test))
```

```python
ts_pred = pd.Series(pred_ar, index=span_test)
```

```python
plt.figure(figsize=(16, 4))
plt.plot(ts_gap_bct.index, ts_gap_bct.values, label="data")
plt.plot(ts_pred.index, ts_pred.values, label="pred")
plt.legend()
plt.show()
```
